# Understanding Figures VS Axes VS Subplots

1. **Figures**:
    - A **figure** in Matplotlib is like a canvas that holds all the visual elements of your plot.
    - It represents the entire plotting area and can contain one or more **axes** (subplots).
    - You create a figure using `plt.figure()` or by using the `fig, ax = plt.subplots()` approach.
    - The figure can have titles, legends, and other global properties.

2. **Axes**:
    - An **axis** (or **subplot**) is a region within the figure where you actually plot your data.
    - Axes define the data coordinate system (x and y axes) and include methods to add annotations like labels, titles, and legends.
    - Each subplot (axis) can have its own data, labels, and styling.
    - Axes are created using methods on the figure object, such as `fig.add_subplot()` or `plt.subplots()`.
    - You can add multiple axes to a single figure, arranging them in rows and columns.

3. **Subplots**:
    - Subplots allow you to display multiple plots (axes) within a single figure.
    - They are organized in a grid layout, specified by the number of rows and columns.
    - You can create subplots using `plt.subplot(rows, columns, index)` or `plt.subplots(rows, columns)`.
    - The `index` parameter determines the position of the current plot within the grid.
    - For example, `plt.subplot(2, 2, 1)` creates a 2x2 grid and places the current plot in the first position.
    - Subplots are useful for comparing related data or visualizing different aspects of your dataset.

In summary:
- **Figures** are the overall canvas that holds everything.
- **Axes** represent individual plots within the figure.
- **Subplots** allow you to arrange multiple axes in a grid layout.

# Axes
An Axes is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects (be aware of the difference between Axes and Axis) that provide ticks and tick labels to provide scales for the data in the Axes. Each Axes also has a title (set via set_title()), an x-label (set via set_xlabel()), and a y-label set via set_ylabel()).

The Axes class and its member functions are the primary entry point to working with the OOP interface, and have most of the plotting methods defined on them (e.g. ax.plot(), shown above, uses the plot method)


````python
import numpy as np
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data) # c = color; s = size
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
ax.legend()
````

# Layouts in Subplot

1. **"constrained"**: This layout ensures that the aspect ratio of each subplot is preserved. It means that if you have multiple subplots, they will be arranged in a way that maintains the same aspect ratio for all of them. This is useful when you want to compare different plots while keeping their proportions consistent.

2. **"auto"**: The default layout, where subplots are evenly spaced and fill the entire figure. The aspect ratio may not be preserved, and the number of rows and columns is determined automatically based on the number of subplots.

3. **"equal"**: Similar to "constrained," but it also ensures that the size of each subplot is the same. This is useful when you want all subplots to have identical dimensions.

4. **"rectangular"**: Subplots are arranged in a rectangular grid, with the specified number of rows and columns. The aspect ratio may not be preserved.

Here's an example of how you can create subplots with the "constrained" layout:

```python
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 1 column
fig, axs = plt.subplots(2, 1, layout='constrained')

# Plot your data in each subplot
axs[0].plot(x_data, y_data1, label='Plot 1')
axs[1].plot(x_data, y_data2, label='Plot 2')

# Customize other plot properties (titles, labels, etc.)

# Show the plots
plt.show()
```
