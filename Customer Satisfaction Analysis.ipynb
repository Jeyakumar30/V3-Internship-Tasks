- Dataset: Customer Personality Analysis
- Task: Use a one-way ANOVA test to compare customer satisfaction scores across different geographical regions identified in the dataset.

```python
import pandas as pd
import random
from scipy.stats import f_oneway
```

```python
data = pd.DataFrame(pd.read_csv("Customer_support_data.csv"))
```

```python
data
```

```python
cities = data["Customer_City"].value_counts()
cities
```

```python
data["CSAT Score"].describe()
```

```python
top_cities = cities[cities.values > 400].index.tolist()
top_cities
```

```python
new_df = data[data["Customer_City"].isin(top_cities)]
new_df.head(5)
```

```python
categories = data.groupby("category").groups.keys()
print("Total no of Groups: ", data.groupby("category").ngroups)
categories
```

```python
dict1 = {}
for category in categories:
    dict1[category] = random.uniform(0, 1)
```

```python
dict1
```

```python
cat_score = pd.DataFrame(pd.Series(dict1))
cat_score.reset_index(inplace=True)
cat_score.rename(columns = {"index":"category", 0:"SAT_Score"}, inplace=True)
cat_score
```

```python
# pd.merge(new_df, cat_score, left_on = new_df["category"].isin(list(categories)),  right_on = cat_score.index.tolist(), how = 'inner')
merged_data = pd.merge(new_df, cat_score, on = "category", how = 'inner')
# merged_data.drop(merged_data.iloc[:,0:])
merged_data
```

```python
merged_data["SAT SCORE"] = merged_data["CSAT Score"] * merged_data["SAT_Score"]
merged_data.columns
```

```python
final = merged_data.drop_duplicates()
```

```python
one_way = final.groupby("Customer_City")["SAT_Score"]
one_way.ngroups
```

```python
for city, group in one_way:
    print(city + " : \n" + str(group))
    print()
```

```python
city_groups = []
for city, group in one_way:
    city_groups.append(group.tolist())
```

```python
f_statistic, p_value = f_oneway(*city_groups)
```

```python
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

#### At 5% level of significance

alpha = 0.05

```python
if p_value < 0.05:
    print("""
    1. Data is statistically significant
    2. We have evidence to support the alternative hypothesis
    3. Null Hypothesis is Rejected.
    
    Conclusion: 
        There is a significant difference in customer satisfaction scores across different geographical regions.
    """)
else:
    print("""
    1. Data is statistically not significant
    2. Null Hypothesis is Accepted.
    
    Conclusion: 
        There is no significant difference in customer satisfaction scores across different geographical regions.
    """)
```
